Python ist eine hochentwickelte, dynamisch typisierte Multiparadigma-Programmiersprache.  
**Beispiel:** Implementierung des klassischen Quicksort-Algorithmus:

````python
def quicksort(arr):                                       # - Funktionsdeklaration mit def
    if len(arr) <= 1:                                     # - if / elif / else             
        return arr                                        # - Einrückung statt geschweifter Klammern!
    pivot  = arr[len(arr) // 2]
    left   = [x for x in arr if x <  pivot]               # Liste als Basistyp, hier bereits als sog. "List Comprehension"
    middle = [x for x in arr if x == pivot]
    right  = [x for x in arr if x >  pivot]
    return quicksort(left) + middle + quicksort(right)


print(quicksort([3, 6, 8, 10, 1, 2, 1]))
````

### Imports

````python
# Standard-Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Besondere Imports für dieses Einsatzszenario
import time
import requests
import urllib
import feedparser
````

### Funktion ```query_arxiv```

````python
def query_arxiv ( search_phrases, fpath_result="abstracts.txt", batchsize=5000, max_batches=50,
                  base_url="http://export.arxiv.org/api/query?", with_dates=False, silent=True ):
    # based on https://static.arxiv.org/static/arxiv.marxdown
    # /0.1/help/api/examples/python_arXiv_parsing_example.txt
    
    # based on the given search_phrases, build the url to connect
    # to the arxiv API
    search_string = "+OR+".join(["abs:%22"+s.replace(" ","+")+"%22" for s in search_phrases])
    url = base_url + "search_query="+search_string+"&max_results="+str(batchsize)
    
    # request data in batches and write to the specified file
    with open(fpath_result,"w") as f:
        for i in range(max_batches):
            # open connection
            data = urllib.request.urlopen(url+"&start="+str(i*batchsize)).read()
            feed = feedparser.parse(data)
            totalresults = int(feed.feed.opensearch_totalresults)
            
            # in the first iteration: report the total number of results; end loop if queue is empty
            # (note that this differs from len(feed.entries) because of the argument max_results in the query
            # which limits each query to the batch size)
            if i == 0:
                if not silent: print("Number of results: "+str(totalresults))
            if len(feed.entries) == 0:
                break
            if not silent: print("Batch "+str(i))
            
            # collect lines in a a buffer and append the buffer to the output file.
            # apply rudimentary masking to special characters.
            buffer = []
            for entry in feed.entries:
                if with_dates:
                    buffer.append( str(entry.published_parsed.tm_year)+"|"
                                  +str(entry.published_parsed.tm_mon)+"|"
                                  +str(entry.published_parsed.tm_mday)+"|" )
                buffer.append ( entry.summary.replace("|",";").replace("\n"," ")+"\n" )
            f.writelines(buffer)

            time.sleep(4) # to adhere to ArXiv terms of use
````

### Funktion ```read_data```

````python
def read_data ( fpath, search_phrases ):
    # read and cleanup; store as a pandas Series for convenience
    with open(fpath,"r") as f:
        abstracts = pd.Series( f.read().splitlines() ).str.lower()
    cleanup = {s:" "+s+" " for s in ["(",")","[","]",",",".",";",":","+","!","?","\"","/","\\","'","~","#"]}
    {
        "(":" ( ",
        ")":" ) ",
        "[":" [ ",
        "]":" ] ",
        ",":" , ",
        ".":" . ",
        ";":" ; ",
        "+":" + ",
        "!":" ! ",
        "?":" ? ",
        "\"":" \" ",
        "'":" ' ",
        "~":""
    }

    # add spaces around special characters; remove multiple spaces
    for k,v in zip(cleanup.keys(), cleanup.values()):
        abstracts = abstracts.str.replace(k, v, regex=False).str.replace("\\s+"," ",regex=True).str.strip()

    # protect search phrases
    for s in search_phrases:
        abstracts = abstracts.str.replace(s, s.replace(" ","_"), regex=False)
    
    # ensure that keywords are kept together: replace spaces by underscores
    search_phrases_protected = [s.replace(" ","_") for s in search_phrases]
    
    return abstracts.tolist() #, search_phrases_protected
````

### Funktion ```document_term_matrix```

````python
def document_term_matrix ( abstracts, batchsize=1000, remove_seldom_quantile=0.0, silent=True ):
    # re-convert to series for convenience
    abstracts = pd.Series(abstracts)
    
    # count word occurencies in each abstract. for memory reasons, do this in batches
    for b in range( int ( np.ceil ( len(abstracts)/batchsize) ) ):
        if not silent: print( "Batch "+str(b))
        _abstracts_countmatrix = abstracts[b*batchsize:(b+1)*batchsize].apply(
            lambda x:pd.Series(x.split("|")[-1].split(" ")).value_counts())#.fillna(0.0).astype(int)

        # throw away very seldom words (unless contained in the search phrases)
        if remove_seldom_quantile > 0:
            idx_keep = _abstracts_countmatrix.sum(axis=0).sort_index()
            idx_keep = idx_keep.index [ (idx_keep > idx_keep.quantile(0.10))
                                       | idx_keep.index.isin(search_phrases_protected) ]
            _abstracts_countmatrix = _abstracts_countmatrix.loc[:,idx_keep].copy()
        
        # in the first iteration, use the temporary matrix as the starting point;
        # in later iterations, concatenate
        if ( b == 0 ):
            abstracts_countmatrix = _abstracts_countmatrix
        else:
            abstracts_countmatrix = pd.concat([abstracts_countmatrix,_abstracts_countmatrix],
                                              axis=0)
    
    # up to now, the "count" for a term that was not found in a document is NA;
    # it is more natural to have zeros here
    abstracts_countmatrix = abstracts_countmatrix.fillna(0.0).astype("int")
    
    return ( abstracts_countmatrix )
````
