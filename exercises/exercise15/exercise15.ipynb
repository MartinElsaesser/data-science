# Teilaufgabe 1
Erzeugen Sie 100 000 Standard-Normalverteilte Zufallsvariablen (vgl. die vorherige Aufgabe) und zeichnen Sie ein Histogramm mit feinem Binning. Zeichnen Sie in dieses Histogramm das 5%, 25%, 50%, 75%, 95%-Quantil ein.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.random.normal(loc=0, scale=1, size=100_000)

quantiles = np.quantile(x, [0.05, 0.25, 0.5, 0.75, 0.95])
df = pd.DataFrame({"quantiles": quantiles, "labels": ["5%", "25%", "50%", "75%", "95%"], "y": np.zeros(quantiles.size)})

plt.hist(x, bins=1000, density=True, alpha=0.5, color='g')
plt.gca().plot(df["quantiles"], df["y"], 'ro', markersize=5)
```


# Teilaufgabe 2
Erzeugen Sie einen Vektor, der 1000 Zahlen zwischen 0 und 10 (jeweils inklusive) "äquidistant" enthält, so dass die Abstände zwischen benachbarten Zahlen genau gleich groß ist.
1. Erstellen Sie in einer gemeinsamen Grafik einen Boxplot, ein Histogramm sowie einen Stripplot.
2. Modifizieren Sie nun den Vektor, so dass die Zahlen nicht mehr gleichmäßig zwischen 0 und 10 verteilt sind; Zahlen dürfen nun nur noch in vier Bereichen liegen, wobei diese vier Bereiche jeweils eine Breite von 0.1 haben dürfen. Führen Sie die Modifikation so durch, dass der Boxplot unverändert bleibt. Erstellen Sie die Grafik aus 2.1) erneut.

```python
import seaborn as sns
vec = np.linspace(0, 10, 1000)
fig, axes = plt.subplots(3,1)

sns.stripplot(data=vec, ax=axes[2])
axes[0].boxplot(vec, vert=False)
axes[1].hist(vec,bins=100)

plt.legend()
```

```python
median = 5
v1 = np.linspace(median+1.5, median+1.4, 250)
v2 = np.linspace(median-1.5, median-1.4, 250)
v3 = np.linspace(median-5, median-4.9, 250)
v4 = np.linspace(median+4.9, median+5, 250)
vec2 = np.concatenate((v1, v2, v3, v4))

fig, axes = plt.subplots(3,1)

axes[0].boxplot(vec2, vert=False)
axes[1].hist(vec2,bins=100)
sns.stripplot(data=vec2, ax=axes[2])

plt.legend()
```
